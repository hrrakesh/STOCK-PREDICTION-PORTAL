```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Types of plots

```python
## line plots
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40 ,50]
plt.plot(x,y);
plt.title("Simple line plot")

plt.ylabel("Y axis label");
plt.xlabel("X axis label")
plt.grid(True)
```

Bar chart

```python
months = ['Jan', 'Feb', 'March', 'May']
revenue = [100, 200, 300, 400]
```

```python
plt.bar(months, revenue)
plt.title("Monthly Revenue Chart")
plt.xlabel("Months")
plt.ylabel("Revenue")
```

## subplots and legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 50]
predicted_price = [8, 21, 25, 28, 48]
```

```python
plt.plot(days, original_price, 'b', label='Original Price'); # codes in single char
plt.plot(days, predicted_price,'y', label='Original Price');
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Days of trading")
plt.ylabel("Price")
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price'); # codes with o for point in single char
plt.plot(days, predicted_price,'r-o', label='Original Price');
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Days of trading")
plt.ylabel("Price")
```

## subplots

```python
## first plot
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title("Original Price over the time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

## second plot
plt.subplot(3, 1, 2) 
plt.plot(days, predicted_price, 'r-o', label='Original Price')
plt.title("Predicted Price over the time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

## Third plot
price_diff = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3, 1, 3) 
plt.bar(days, price_diff, color='purple', label='Price diff')
plt.title(" Price diff over the time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

plt.tight_layout()
## adds padding to all plot
```

## saving the result 

```python
plt.plot(days, original_price, 'b-o', label='Original Price'); # codes with o for point in single char
plt.plot(days, predicted_price,'r-o', label='Original Price');
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Days of trading")
plt.ylabel("Price")
plt.savefig("Example-fig.png", dpi=300)
```
